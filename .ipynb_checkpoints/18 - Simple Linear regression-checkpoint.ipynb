```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('height-weight.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Weight'],df['Height'])
```

```python
## Correlation
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
# Independent and dependent features
X=df[['Weight']] #independent feature should be DataFrame or 2D array
y=df['Height'] # Dependent varaibale can be in series or 1D
np.array(X).shape
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Apply Simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression(n_jobs=-1)
```

```python
regression.fit(X_train, y_train)
```

```python
print("Coefficient or slope:",regression.coef_)
print("Interpret:",regression.intercept_)
```

```python
#Training data best fit line
plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train))
```

### Prediction of test data
1. predicted height output = intercept + coef_(Weights)
2. y_pred_test = 166.74 + 10.26(X_test)

```python
# Predict for the test data
regression.predict(X_test)
```
